```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import dct, idct
from PIL import Image
import os
```

```python
# Load the image from the current directory
image_path = 'download.jpg'  # Replace with your image file name
image = Image.open(image_path)

# Convert the image to a numpy array
image_array = np.array(image)

# Display the original image
plt.imshow(image_array)
plt.title("Original Image")
plt.axis('off')
plt.show()

# Display the image dimensions and size
print(f"Original Image Dimensions: {image_array.shape}")
print(f"Original Image Size: {os.path.getsize(image_path) / 1024:.2f} KB")
```

```python
# Function to apply DCT and then inverse DCT
def apply_dct_compression(image_array, compression_ratio=0.5):
    # Convert to float for DCT computation
    image_float = image_array.astype(float)
    
    # Apply DCT to each channel (R, G, B)
    dct_transformed = np.zeros_like(image_float)
    for i in range(3):  # Assuming RGB channels
        dct_transformed[:, :, i] = dct(dct(image_float[:, :, i], axis=0, norm='ortho'), axis=1, norm='ortho')
    
    # Zero out small coefficients for compression
    cutoff = int(dct_transformed.shape[0] * compression_ratio)
    dct_transformed[cutoff:, cutoff:, :] = 0
    
    # Apply inverse DCT to reconstruct the image
    idct_reconstructed = np.zeros_like(dct_transformed)
    for i in range(3):
        idct_reconstructed[:, :, i] = idct(idct(dct_transformed[:, :, i], axis=0, norm='ortho'), axis=1, norm='ortho')
    
    # Clip values to the valid range (0-255) and convert back to uint8
    compressed_image_array = np.clip(idct_reconstructed, 0, 255).astype(np.uint8)
    
    return compressed_image_array

# Apply DCT compression
compressed_image_array = apply_dct_compression(image_array, compression_ratio=0.1)

# Display the compressed image
plt.imshow(compressed_image_array)
plt.title("Compressed Image")
plt.axis('off')
plt.show()

# Display the compressed image dimensions and estimate its size
compressed_image = Image.fromarray(compressed_image_array)
compressed_image_path = 'compressed_image.jpg'
compressed_image.save(compressed_image_path)
print(f"Compressed Image Dimensions: {compressed_image_array.shape}")
print(f"Compressed Image Size: {os.path.getsize(compressed_image_path) / 1024:.2f} KB")
```

```python
# Get the size of the original and compressed images in KB
original_image_size = os.path.getsize(image_path) / 1024  # Size in KB
compressed_image_size = os.path.getsize(compressed_image_path) / 1024  # Size in KB

# Display the original and compressed images side by side
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Original image
axes[0].imshow(image_array)
axes[0].set_title(f"Original Image\nSize: {original_image_size:.2f} KB")
axes[0].axis('off')

# Compressed image
axes[1].imshow(compressed_image_array)
axes[1].set_title(f"Compressed Image\nSize: {compressed_image_size:.2f} KB")
axes[1].axis('off')

plt.show()
```
